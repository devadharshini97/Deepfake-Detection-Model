```python
from facenet_pytorch import MTCNN
import cv2
from PIL import Image
from os import listdir, makedirs
import glob
from os.path import join, exists
from skimage.io import imsave
import imageio.core.util
import os
```

```python
def ignore_warnings(*args, **kwargs):
    pass

imageio.core.util._precision_warn = ignore_warnings
```

```python
mtcnn = MTCNN(
    margin=40,
    select_largest=False,
    post_process=False,
    device="cuda:0"
)
```

```python
source_frames_folders = ["D:/sem2/adv_ML/Project/dataset/frames2Real4"]
```

```python
source_frames_folders
```

```python
dest_faces_folder = "D:/sem2/adv_ML/Project/dataset/faces/0/"
```

```python
for i in source_frames_folders:
    counter = 0
    for j in listdir(i):
        imgs = glob.glob(join(i, j, "*.jpg"))
        if counter % 1000 == 0:
            print("Number of videos done:", counter)
        if not exists(join(dest_faces_folder, j)):
            makedirs(join(dest_faces_folder, j))
            #os.chmod(dest_faces_folder, 0o777)
        for k in imgs:
            frame = cv2.imread(k)
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frame = Image.fromarray(frame)
            face = mtcnn(frame)
            x = face.permute(1,2,0).int().numpy()
            try:
                imsave(
                    join(dest_faces_folder, j, k.split("\n")[-1]),
                    face.permute(1, 2, 0).int().numpy()
                )
            except AttributeError:
                print("Image skipping")
        counter += 1
```
