```python
import numpy as np
from keras.utils import load_img, img_to_array
import pandas as pd
from keras_facenet import FaceNet

# Read csv file contain image face paths
data = pd.read_csv("/content/drive/MyDrive/train_faces_25frames.csv")

images = data["images_list"]
labels = data["label"]

train_data = []
train_label = []
count = 0

embedder = FaceNet()
```

```python
for (img_path, label) in zip(images,labels):
  x = load_img(img_path) #(160, 160) size of x
    # x = img_to_array(img_path)
  #print(x.size)
  x = np.expand_dims(x, axis=0) #shape of x is (1, 160, 160, 3)
  #print(x.shape)
  embeddings = embedder.embeddings(x)
  train_data.append(embeddings)
  train_label += [label]

  if count % 50 == 0:
      print("Number of files done:", count)
  count += 1

train_data = np.array(train_data)
train_label = np.array(train_label)

np.save("/content/drive/MyDrive/train_data_facenet_embeddings.npy", train_data)
np.save("/content/drive/MyDrive/train_label_facenet_embeddings.npy", train_label)
print("Files saved....")
```
